# Code Helper AI

Provide a code that you are not sure of, and AI will provide a detailed explanation.

````python
# imports

from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display
from openai import OpenAI
import ollama
````

````python
# constants

MODEL_GPT = 'gpt-4o-mini'
````

````python
# set up environment

load_dotenv()
openai = OpenAI()
````

````python
my_question = input("Please ask a question: ")
````

````python
# prompts

system_prompt = "You are a helpful technical tutor who answers questions about python code, SQL code, software engineering, data science and LLMs"
user_prompt = "Please give a detailed explanation to the following question: " + my_question
````

````python
# messages

messages = [
    {'role': 'system', 'content': system_prompt},
    {'role': 'user', 'content': user_prompt}
]
````

````python
# Get gpt-4o-mini to answer, with streaming

stream = openai.chat.completions.create(model=MODEL_GPT, messages=messages, stream=True)

response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    response = response.replace("```","").replace("markdown", "")
    update_display(Markdown(response), display_id=display_handle.display_id)
````
